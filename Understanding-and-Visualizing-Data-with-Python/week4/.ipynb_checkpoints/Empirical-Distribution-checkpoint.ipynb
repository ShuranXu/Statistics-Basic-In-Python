```python
import warnings
warnings.filterwarnings('ignore')
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.distributions.empirical_distribution import ECDF
```

```python
# set the random seeds
random.seed(1738)
```

```python
# set the distribution mean and std
mu = 7
sigma =1.7

# normalvariate() is an inbuilt method of the random module. 
# It is used to return a random floating point number with normal distribution.
Observations = [random.normalvariate(mu, sigma) for _ in range(1000)]
Observations
```

```python
# draw the distribution using a histogram plot
plt.figure();
plt.hist(Observations)

# Add a vertical line to signify 1 * std_dev
plt.axvline(np.mean(Observations) - np.std(Observations), color= 'g')
plt.axvline(np.mean(Observations) + np.std(Observations), color = 'g')

plt.axvline(np.mean(Observations) + (np.std(Observations) * 2), color = "y")
plt.axvline(np.mean(Observations) - (np.std(Observations) * 2), color = "y")
```

```python
# Randomly choose a list of 100 numbers extracted from the Observation using random.sample()
sampleA = random.sample(Observations, 100)
sampleB = random.sample(Observations, 100)
sampleC = random.sample(Observations, 100)
```

```python
# plot a histogram for each sample list
fig, ax = plt.subplots()
sns.distplot(a=sampleA, ax=ax, bins=10)
sns.distplot(a=sampleB, ax=ax, bins=10)
sns.distplot(a=sampleC, ax=ax, bins=10)
```

```python
'''
The empirical distribution/empirical CDF is a cumulative density function that signifies the proportion 
of observations that are less than or equal to a certain values.

Note that the cumulative distribution function (CDF) is a distribution function that indicates the 
probability that X will take a value less than or equal to x
'''
ecdf = ECDF(Observations)

plt.plot(ecdf.x, ecdf.y)
plt.axhline(y = 0.025, color = 'y', linestyle='-')
plt.axvline(x = np.mean(Observations) - (2 * np.std(Observations)), color = 'y', linestyle='-')

plt.axhline(y = 0.975, color = 'y', linestyle='-')
plt.axvline(x = np.mean(Observations) + (2 * np.std(Observations)), color = 'y', linestyle='-')
```
